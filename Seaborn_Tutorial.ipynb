```python
# Pandas for managing datasets
import pandas as pd
```

```python
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python

# Display first 5 observations
df.head()
```

Scatter Plot

```python

# Scatterplot arguments
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=True, # No regression line
           hue='Stage')   # Color by evolution stage
```

Box Plots

```python
# Boxplot
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=df)
```

```python

# Preprocess DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

# New boxplot using stats_df
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=stats_df)
```

Violin Plot

```python
# Violin plot
plt.figure(figsize=(12,8)) # Set plot dimensions
sns.violinplot(x='Type 1', y='Attack', data=df)
```

Heat Map

```python
# Select only the relevant numerical features
num_features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
num_df = df[num_features]

# 1. Pearson Correlation Coefficient
pearson_corr = num_df.corr(method='pearson')

# 2. Spearman Rank Correlation
spearman_corr = num_df.corr(method='spearman')

# 3. Plot Pearson Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Pearson Correlation Heatmap", fontsize=14)
plt.show()

# 4. Plot Spearman Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Spearman Correlation Heatmap", fontsize=14)
plt.show()
```
